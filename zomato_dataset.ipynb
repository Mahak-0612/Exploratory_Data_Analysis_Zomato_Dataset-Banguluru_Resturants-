Removing "New", "-" & "/5" from Rate Column:

```python
def handlingrate(value):
    if value == 'NEW' or value == '-':
        return np.nan
    else:
        value = str(value).split("/")
        value = value[0]
        return float(value)

df['rate'] = df['rate'].apply(handlingrate)
df['rate'].head()
```

Filling Null Values in Rate Column with Mean

```python
df.rate.isnull().sum()
```

```python
df['rate'].fillna(df['rate'].mean(), inplace = True)
df['rate'].isnull().sum()
```

```python
df.info()
```

```python
# we can notice some values are containing comma which restricts it from becoming an integer or float, therefore we'll create a function to remove the sam
```

Removing , from Cost2Plates Column

```python
def handlecomma(value):
    value = str(value)
    if ',' in value:
        value = value.replace(',', '')
        return float(value)
    else:
        return float(value)

df['Cost2plate'] = df['Cost2plate'].apply(handlecomma)
df['Cost2plate'].unique()
```

```python
df.head()
```

Cleaning Rest Type Column:

```python
df['rest_type'].value_counts()
```

```python
rest_type = df['rest_type'].value_counts(ascending = False)
rest_type
```

```python
rest_type_lessthan1000 = rest_type[rest_type<1000]
rest_type_lessthan1000
```

Making Rest Types less than 1000 in frequency as others:

```python
def handle_rest_type(value):
    if value in rest_type_lessthan1000:
        return 'others'
    else:
        return value

df['rest_type'] = df['rest_type'].apply(handle_rest_type)
df['rest_type'].value_counts()
```

```python
df.head()
```

Cleaning Location Column:

```python
df['location'].value_counts()
```

```python
location = df['location'].value_counts(ascending=False)
location_lessthan300 = location[location < 300]

def handle_location(value):
    if value in location_lessthan300:
        return 'others'
    else:
        return value

df['location'] = df['location'].apply(handle_location)
df['location'].value_counts()
```

Cleaning Cuisines Columns:

```python
cuisines = df['cuisines'].value_counts(ascending=False)
cuisines_lessthan100 = cuisines[cuisines < 100]

def handle_cuisines(value):
    if value in cuisines_lessthan100:
        return 'others'
    else:
        return value

df['cuisines'] = df['cuisines'].apply(handle_cuisines)
df['cuisines'].value_counts()
```
